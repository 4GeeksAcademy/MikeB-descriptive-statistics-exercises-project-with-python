### Exercise 1

We will use NumPy to obtain information to describe statistically.

- Generate an array of 100 elements following a normal distribution.
- Generate an array of 100 elements following a chi-square distribution with 3 degrees of freedom.
- Calculate the main metrics and statistical measures that best describe the two vectors.

````python
# TODO
import random
import numpy as np
import statistics as stats
import matplotlib.pyplot as plt
from scipy.stats import skew
from scipy.stats import kurtosis

normDistArr = np.random.normal(size=100)
chisqrDistArr = np.random.chisquare(df=3, size=100)

print(f"Normal Data: {normDistArr}")
print()

print(f"Chi-Square Data: {chisqrDistArr}")
print()

# Mean
normMean = stats.mean(normDistArr)
chiMean = stats.mean(chisqrDistArr)

print(f"The Mean of Normal: {normMean}")
print(f"The Mean of Chi-square: {chiMean}")
print()


# Median
normMedian = stats.median(normDistArr)
chiMedian = stats.median(chisqrDistArr)

print(f"The Median of Normal: {normMedian}")
print(f"The Median of Chi-square: {chiMedian}")
print()


# Mode
normMode = stats.mode(normDistArr)
chiMode = stats.mode(chisqrDistArr)

print(f"The Mode of Normal: {normMode}")
print(f"The Mode of Chi-square: {chiMode}")
print()


# Range
ranNormal = max(normDistArr) - min(normDistArr)
ranChiSqr = max(chisqrDistArr) - min(chisqrDistArr)

print(f"The range of Normal: {ranNormal}")
print(f"The range of Chi-square: {ranChiSqr}")
print()


# Variance and Standar Deviation
varNormal = stats.variance(normDistArr)
stdNormal = stats.stdev(normDistArr)
varChiSqr = stats.variance(chisqrDistArr)
stdChiSqr = stats.variance(chisqrDistArr)
stdChiSqr = stats.stdev(chisqrDistArr)

print(f"The Variance of Normal: {varNormal}")
print(f"The Standard Deviation of Normal: {stdNormal}")
print(f"The Variance of Chi-square: {varChiSqr}")
print(f"The Standard Deviation of Chi-square: {stdChiSqr}")
print()


# Skew
skewNormal =skew(normDistArr)
skewChiSqr = skew(chisqrDistArr)

print(f"The Skew of Normal: {skewNormal}")
print(f"The Skew of Chi-square: {skewChiSqr}")
print()

# Skew
kurtosisNormal = kurtosis(normDistArr)
kurtosisChiSqr = kurtosis(chisqrDistArr)

print(f"The Kurtosis of Normal: {kurtosisNormal}")
print(f"The Kurtosis of Chi-square: {kurtosisChiSqr}")

# Plot
plt.hist(normDistArr, bins = 30, alpha = 0.5, density=True, label = "Normal Distribution")
plt.hist(chisqrDistArr, bins = 30, alpha = 0.5, density=True, label = "Chi-Square Distribution")
plt.legend()
plt.title('Descriptive Statistics')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
````

### Exercise 2

Write a Python program to calculate the standard deviation of the following data:

```py
data = [4, 2, 5, 8, 6]
```

````python
# TODO
import statistics as stats

data = [4, 2, 5, 8, 6]

total = 0
for i in range (len(data)):
    total += data[i]

mean = total / len(data)

sqrDiff = []
for i in range (len(data)):
    sqrDiff.append((data[i]-mean)**2)

sumSqrDiff = 0
for i in range (len(sqrDiff)):
    sumSqrDiff += sqrDiff[i]

stdDevCalc = (sumSqrDiff / len(data))**0.5
print(stdDevCalc)

# # With the Python build in functions comparison
# std = stats.stdev(data)
# print(std)
````
